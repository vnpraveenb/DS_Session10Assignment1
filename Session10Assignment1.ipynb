```python
#Read the dataset from the below link
#   https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv

import numpy as np
import pandas as pd

baby_names = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")
baby_names.head()
```

```python
#Questions:
   
# 1. Delete unnamed columns
baby_names.drop(baby_names.columns[baby_names.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
baby_names.head()
 
```

```python
# 2. Show the distribution of male and female
baby_names['Gender'].value_counts()
```

```python
# 3. Show the top 5 most preferred names ( Showing the names sorted in descending order( Most to least))

preferred_names = baby_names.groupby("Name").sum()
preferred_names.sort_values("Count",ascending=0).head()
```

```python
#4. What is the median name occurence in the dataset

# for Ascending order of names which have same occurence as median()

preferred_names[preferred_names.Count == preferred_names.Count.median()]

#       or

# for Descending order of names which have same occurence as median()

# preferred_names.sort_values("Count",ascending=0,inplace=True) 
# preferred_names[preferred_names.Count == preferred_names.Count.median()] 

```

```python
#5. Distribution of male and female born count by states
 
baby_names.groupby(['State', 'Gender']).Count.sum()
```
